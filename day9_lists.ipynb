```python
def is_anagram(s1,s2):
    """Check the anagram"""
    return sorted(s1.lower()) == sorted(s2.lower())
```

```python
is_anagram('stake','skate')
```

```python
def reversed_word(word):
    """Reverse the word"""
    return ''.join(reversed(word))
```

```python
def is_palindrome(word):
    """To check the word is palindrome."""
    if word == reversed_word(word):
        return True
    return False
```

```python
is_palindrome('noon')
```

```python
is_palindrome('robert')
```

```python
word_list = ['tomato','noon','rotator']
for word in word_list:
    if len(word) >=7 and is_palindrome(word):
        print(word)
```

```python
"""Write a function called reverse_sentence that takes as an argument a string that
contains any number of words separated by spaces. It should return a new string that
contains the same words in reverse order. For example, if the argument is “Reverse
this sentence,” the result should be “Sentence this reverse.”
Hint: you can use the capitalize methods to capitalize the first word and convert
the other words to lowercase."""
```

```python
def reverse_sentence(sentence):
    words = sentence.split()
    reversed_words = words[::-1]

    reversed_words[0] = reversed_words[0].capitalize()
    for i in range(1,len(reversed_words)):
        reversed_words[i] = reversed_words[i].lower()

    return ' '.join(reversed_words)
```

```python
reverse_sentence("Reverse this sentence")
```
