```python
import unicodedata
```

```python
unicodedata.category('A')
```

```python
unicodedata.category('"')
```

```python
punc = {}
for line in open(filename):
    for char in line:
        category1 = unicodedata.category(char)
        if category1.startswith('P'):
            punc[char] = 1
```

```python
punctuation = ''.join(punc)
print(punctuation)
```

```python
def clean_word(word):
    return word.strip(punctuation).lower()
```

```python
clean_word('Behold,,,;;::')
```

```python
unique_words2 = {}
for line in open(filename):
    for word in split_line(line):
        word = clean_word(word)
        unique_words2[word] = 1

len(unique_words2)
```

```python
sorted(unique_words2, key=len)[-5:]
```

```python
word_counter = {}
for line in open(filename):
    for word in split_line(line):
        word = clean_word(word)
        if word not in word_counter:
            word_counter[word] = 1
        else:
            word_counter[word] += 1
```

```python
def second_element(t):
    return t[1]
```

```python
items = sorted(word_counter.items(), key=second_element, reverse=True)
```

```python
for word, freq in items[:5]:
    print(freq, word)
```

```python
def print_most_common(word_counter, num = 5):
    items = sorted(word_counter.items(), key=second_element, reverse=True)
    for word, freq in items[:num]:
        print(freq, word, sep='\t')
```

```python
print_most_common(word_counter)
```

```python
print_most_common(word_counter, 3)
```

```python
word_list = open('words.txt').read().split()
```

```python
valid_words = {}
for word in word_list:
    valid_words[word] = 1
```

```python
def substract(d1, d2):
    res = {}
    for key in d1:
        if key not in d2:
            res[key] = d1[key]
    return res
```

```python
diff = substract(word_counter, valid_words)
```

```python
print_most_common(diff)
```

```python
actual_mispelled = []
for word, freq in diff.items():
    if freq == 1:
        actual_mispelled.append(word)
```
