```python
def find_first(pattern):
    for line in open('pg345.txt',encoding='utf-8'):
        result = re.search(pattern,line)
        if result != None:
            return result
```

```python
result = find_first('Harker')
result.string
```

```python
def count_matches(pattern):
    count = 0
    for line in open('pg345.txt',encoding='utf-8'):
        result = re.search(pattern,line)
        if result != None:
            count += 1
    return count
```

```python
count_matches('Mina|Murray')
```

```python
# Wordle
not_in_word = 'TID'
def check_word(word):
    if 'R' not in word:
        return False
    if 'R' == word[4]:
        return False

    if 'E' not in word:
        return False
    for letter in not_in_word:
        if letter in word:
            return False
    return True        
```

```python
print(check_word('MOWER'))
```
